```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
import pandas as pd
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier, Pool
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv('train.csv')
```

```python
df['transaction_time'] = pd.to_datetime(df['transaction_time'])
df['minutes'] = df['transaction_time'].dt.minute
df['hours'] = df['transaction_time'].dt.hour
df['days'] = df['transaction_time'].dt.day
df['month'] = df['transaction_time'].dt.month
df['year'] = df['transaction_time'].dt.year

df['hour_sin'] = np.sin(2 * np.pi * df['hours'] / 24)
df['hour_cos'] = np.cos(2 * np.pi * df['hours'] / 24)

df['day_sin'] = np.sin(2 * np.pi * df['days'] / 30) 
df['day_cos'] = np.cos(2 * np.pi * df['days'] / 30)

df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)
```

```python
df = df.drop('transaction_time', axis=1)
df = df.drop(['lon', 'lat', 'post_code'], axis=1)
```

```python
X = df.drop('target', axis=1)
y = df['target']

cat_features = ['merch', 'cat_id', 'name_1', 'name_2', 'gender', 'street', 'one_city', 'us_state', 'jobs']

full_pool = Pool(X, label=y, cat_features=cat_features)

model = CatBoostClassifier(iterations=300, verbose=50) 
model.fit(full_pool)
```

```python
df_test = pd.read_csv('test.csv')
df_test
```

```python
df_test['transaction_time'] = pd.to_datetime(df_test['transaction_time'])

df_test['minutes'] = df_test['transaction_time'].dt.minute
df_test['hours'] = df_test['transaction_time'].dt.hour
df_test['days'] = df_test['transaction_time'].dt.day
df_test['month'] = df_test['transaction_time'].dt.month
df_test['year'] = df_test['transaction_time'].dt.year

df_test['hour_sin'] = np.sin(2 * np.pi * df_test['hours'] / 24)
df_test['hour_cos'] = np.cos(2 * np.pi * df_test['hours'] / 24)

df_test['day_sin'] = np.sin(2 * np.pi * df_test['days'] / 30) 
df_test['day_cos'] = np.cos(2 * np.pi * df_test['days'] / 30)

df_test['month_sin'] = np.sin(2 * np.pi * df_test['month'] / 12)
df_test['month_cos'] = np.cos(2 * np.pi * df_test['month'] / 12)

df_test = df_test.drop('transaction_time', axis=1)
df_test = df_test.drop(['lon', 'lat', 'post_code'], axis=1)
```

```python
# df_test
```

```python
y_predict = model.predict(df_test)
```

```python
submission = pd.DataFrame({'index': df_test.index, 'prediction': y_predict})
submission.set_index('index', inplace=True)

submission.to_csv('submission.csv')

print("Файл submission.csv успешно создан.")
```
