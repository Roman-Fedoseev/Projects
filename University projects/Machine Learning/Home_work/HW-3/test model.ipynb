```python
import numpy as np
import pandas as pd
from IPython.display import display
from sklearn.metrics import fbeta_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

### Область работы 2 (выполнение лучшей модели)

```python
# Данный блок(и) в области 2 выполняется преподавателем
#
# В области находится одна, единственная, итоговая модель машинного обучения с однозначными, 
# зафиксированными параметрами
#
# В данной области категорически запрещается искать, выбирать, улучшать, оптимизировать, 
# тюниговать и т.д. модель машинного обучения
```

```python
# Путь к тренировочному набору
path_train = 'train.csv' 
path_y = 'train_Y.csv'
# Путь к тестовому набору
path_test  = 'test.csv' 
```

```python
X_df = pd.read_csv(path_train)
y_df = pd.read_csv(path_y)
df_test = pd.read_csv(path_test)

df = X_df.join(y_df, how='outer')
df = df.drop_duplicates()

X = df.drop('Diabetes_binary', axis=1)
y = df['Diabetes_binary']

class_weights = {0: 1, 1: 2.9}
sample_weight = np.array([class_weights[i] for i in y])
```

```python
# Блок предсказания с использованием тестового набора (все три модели)
gb = GradientBoostingClassifier(n_estimators=300, max_depth=4)
gb.fit(X, y, sample_weight=sample_weight)
```

```python
rf = RandomForestClassifier(n_estimators=1000, n_jobs = -1, criterion= 'gini', max_features='sqrt', max_depth=10, min_samples_split=9)
rf.fit(X, y, sample_weight=sample_weight)
```

```python
# Вектора предсказанных значений  RandomForestClassifier или ExtraTreesClassifier
y_predict_RF = rf.predict(df_test)
```

```python
# Вектора предсказанных значений  GradientBoostingClassifier или HistGradientBoostingClassifier
y_predict_GB = gb.predict(df_test)
```

```python
# Вектора предсказанных значений  финальной модели
y_predict_final = gb.predict(df_test)
```
